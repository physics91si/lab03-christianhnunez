```python
# Imports
import string
# Don't edit this function
def load_sample():
    """Return the entire text of Hamlet in a string."""
    with open('hamlet.txt') as f:
        sample = f.read()
    return sample
```

```python
def parse_sample(sample, words_only=False, sort_list=False):
    sample = sample.replace('-', ' ')
    if(words_only): 
        sample = sample.lower()
        sample = sample.translate(None, string.punctuation)
    wordArray = sample.split() #Array of tokens
    if(sort_list):
        wordArray.sort()   
    return wordArray
    
```

```python
print(parse_sample(load_sample(), True, True))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def count_freq(lst):
    """Return a dict of word frequencies given a list of words.
    
    Arguments:
    -- lst: a list of strings
    
    Returns:
    -- freq_dict: a dictionary of word: frequency pairs. The keys are strings
       containing the words, and the values are integers indicating how many
       times each word appears.
    
    Example:
    >>> count_freq(['time', 'after', 'time'])
    {'time': 2, 'after': 1}
    """
    freq_dict = {}
    
    for i in range(0, len(lst)):
        if lst[i] not in freq_dict:
            freq_dict[ lst[i] ] = 1
        else:
            freq_dict[ lst[i] ] += 1
    
    return freq_dict

def mean(lst):
    return np.mean(lst)

def stdev(lst):
    return np.std(lst)

def median(lst):
    return np.median(lst)

def plot_hist(lst):
    plt.hist(lst, bins=100, range = (0,50))
    plt.xlabel("word frequency")
    plt.ylabel("# of words")
    plt.show()
    
    
```

Part 3
------

```python
def print_stats(lst):
    freq_dict = count_freq(lst)
    freq_list = freq_dict.values()
    print("The mean word frequency is: " + str(mean(freq_list)))
    print("The standard deviation of the frequency distribution is: " + str(stdev(freq_list)))
    print("The median word frequency is: " + str(median(freq_list)))
    print("")
    plot_hist(freq_list)
    
```

```python
print_stats(parse_sample(load_sample(), True, True))
```
